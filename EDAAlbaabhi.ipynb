# outliers

```python
sns.boxplot(x='price',data=df)
plt.title('box plot')
```

```python
Q1,Q3=df['price'].quantile([.25,.75])
IQR= Q3-Q1
LL=Q1-1.5*(IQR)
UR=Q3+1.5*(IQR)
```

```python
print(LL,UR)
```

```python
def limit_imputer(x):
    if x>UR:
        return UR
    if x<LL:
        return LL
    else:
        return x
    
```

```python
df['price'] =df['price'].apply(limit_imputer)
```

```python
sns.boxplot(x='price',data=df)
```

# Q10. Check the correlation between curb-weight and Miles per gallon('city-mpg','highway-mpg')? Also calculate the correlation Coefficient

```python
sns.regplot(x='curb-weight',y='city-mpg',data=df)
plt.title('correlation B/W curb- weight & city-mpg')
plt.show()

sns.regplot(x='curb-weight',y='highway-mpg',data=df)
plt.title('correlation B/W curb- weight & highway-mpg')
plt.show()
```

```python
corr_city = df['curb-weight'].corr(df['city-mpg'])
corr_highway = df['curb-weight'].corr(df['highway-mpg'])
print('Correlation coefficient (Curb Weight vs. City MPG):', corr_city)
print('Correlation coefficient (Curb Weight vs. Highway MPG):', corr_highway)
```
